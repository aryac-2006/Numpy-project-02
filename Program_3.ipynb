```python
import numpy as np

# Create a 4 5 matrix of random integers between 20 and 80
arr = np.random.randint(20, 80, (4, 5))
print("Matrix of(4x5): ")
print(arr)

# Minimum and maximum value
min_value = np.min(arr)
max_value = np.max(arr)
print("\n1) Minimum value: ", min_value)
print("2) Maximum value: ", max_value)

# Sum of all elements
sum = np.sum(arr)
print("3) Sum of all elements: ", sum)

# Mean and Standard Deviation
m = np.mean(arr)
print("4) Mean: ", m)
std = np.std(arr)
print("5) Standard Deviation: ", std)
```
