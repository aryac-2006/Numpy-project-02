```python
import numpy as np

arr = np.random.randint(1, 51, 20)

print("1D Array: ")
print(arr)

min_value = np.min(arr)
min_index = np.argmin(arr)
print("\n1) Minimum value:- ", min_value, ",  Index:- ", min_index)

max_value = np.max(arr)
max_index = np.argmax(arr)
print("2) Maximum value:- ", max_value, ",  Index:- ", max_index)

sum = np.sum(arr)
print("3) Sum of all elements: ", sum)

mean = np.mean(arr)
print("4) Mean: ", mean)

std = np.std(arr)
print("5) Standard Deviation: ", std)
```
