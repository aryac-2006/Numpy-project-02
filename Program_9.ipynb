```python
import numpy as np

# Generate a 6x6 matrix of random numbers
matrix = np.random.randn(6, 6)

# Print the matrix
print("Original Matrix:")
print(matrix)

# Print shape, size, and data type
print("\nShape:", matrix.shape)
print("Size:", matrix.size)
print("Data Type:", matrix.dtype)

# Find the index of the maximum and minimum value
print("\nIndex of Maximum Value:", np.argmax(matrix))
print("Index of Minimum Value:", np.argmin(matrix))

# Extract the top-left 3x3 submatrix
print("\nTop-Left 3x3 Submatrix:")
print(matrix[:3, :3])

# Replace all negative numbers with their absolute value
matrix[matrix < 0] = np.abs(matrix[matrix < 0])

# Print the modified matrix
print("\nModified Matrix:")
print(matrix)

# Print the mean of the modified matrix
print("\nMean of Modified Matrix:")
print(np.mean(matrix))
```
