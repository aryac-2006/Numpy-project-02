```python
import numpy as np

# Create a 1D array containing numbers from 1 to 24 using np.arange()
a = np.arange(1, 25)
arr = np.array(a)

print("1) Original 1D Array: ")
print(arr)
print("Shape: ", arr.shape)

# Reshape it into (4, 6)
arr2 = np.array(a).reshape(4, 6)
print("\n2) Reshaped 2D Array: ")
print(arr2)
print("Shape: ", arr2.shape)

# Reshape it into (3, 8)
arr3 = np.array(a).reshape(3, 8)
print("\n3) Reshaped 2D Array: ")
print(arr3)
print("Shape: ", arr3.shape)

# Reshape it into (2, 3, 4)
arr4 = np.array(a).reshape(2, 3, 4)
print("\n4) Reshaped 3D Array: ")
print(arr4)
print("Shape: ", arr4.shape)

```
