```python
import numpy as np

a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

print("1D Array: ")
print("a = ", a)
print("b = ", b)

# Addition
print("\n1) Addition: ")
print(a + b)

# Subtraction
print("\n2) Subtraction: ")
print(a - b)

# Multiplication
print("\n3) Multiplication: ")
print(a * b)

# Division
print("\n4) Division: ")
print(a / b)

# Element-wise power
print("\n5) Element-wise power: ")
print(a ** b)

# Modulo operation %
print("\n6) Modulo operation: ")
print(a % b)
```
