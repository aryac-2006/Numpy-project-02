```python
import numpy as np

# Generate a 2D array of shape (10, 5) with random marks between 30 and 100
marks = np.random.randint(30, 101, (10, 5))

print("=== Student Marks Analyzer ===")
# Print the marks array
print("\nStudent Marks (10 Students × 5 Subjects):")
print(marks)

# Calculate total marks of each student
total = np.sum(marks, axis=1)
print("\nTotal Marks of Each Student:")
print(total)

# Calculate average marks of each student
average = np.mean(marks, axis=1)
print("\nAverage Marks of Each Student:")
print(average)

# Find the student with highest total marks
highest = np.argmax(total)
print("\nStudent with Highest Total Marks:", highest)
print("Highest Total Marks:", total[highest])

# Find the student with lowest total marks
lowest = np.argmin(total)
print("\nStudent with Lowest Total Marks:", lowest)
print("Lowest Total Marks:", total[lowest])

# Calculate overall class mean
class_mean = np.mean(marks)
print("\nOverall Class Mean:", class_mean)

# Calculate overall class standard deviation
class_std = np.std(marks)
print("Overall Class Standard Deviation:", class_std)

# Reshape the array into 5 × 10
reshaped = marks.reshape(5, 10)
print("\nReshaped Array (5 × 10):")
print(reshaped)

# Find the indices of the top 3 students based on total marks
top3 = np.argsort(total)[-3:][::-1]

# Extract marks of the top 3 students
print("\nTop 3 Students' Marks:")
print(marks[top3])
```
